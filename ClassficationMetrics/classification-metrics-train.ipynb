```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.sample(10)
```

```python
df['label'].nunique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
log = LogisticRegression()
log.fit(X_train,y_train)
y_pred1 = log.predict(X_test)
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
y_pred2 = dtree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of Logistic Regression",accuracy_score(y_test,y_pred1))
print("Accuracy of Decision Trees",accuracy_score(y_test,y_pred2))
```

```python
print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,10)))
```

```python
print("Decision Tree Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,10)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred2))
```
