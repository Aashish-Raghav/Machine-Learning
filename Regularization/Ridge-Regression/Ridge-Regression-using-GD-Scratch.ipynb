```python
from sklearn.datasets import load_diabetes
import numpy as np
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

# SGDRegressor

```python
reg = SGDRegressor(penalty='l2',max_iter=500,learning_rate='constant',eta0=0.1,alpha=0.0001)
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
print(reg.coef_,reg.intercept_)
```

# Ridge scikit learn

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=0.001, max_iter=500,solver='sparse_cg')
```

```python
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

# Mera Ridge

```python
class MeraRidge:
    def __init__(self,epochs,learning_rate,alpha):
        self.lr = learning_rate
        self.epochs = epochs
        self.alpha = alpha
        self.coef_ = None
        self.intercept_ = None

    def fit(self,X,y):
        X = np.insert(X,0,1,axis=1)
        thetha = np.ones(X.shape[1])
        thetha[0] = 0
        for i in range(self.epochs):
            thetha_der = np.dot(X.T,X).dot(thetha) - np.dot(X.T,y) + self.alpha*thetha
            thetha = thetha - self.lr*thetha_der

        self.coef_ = thetha[1:]
        self.intercept_ = thetha[0]

    def predict(self,X):
        return np.dot(X,self.coef_) + self.intercept_
```

```python
Ridge = MeraRidge(epochs=500,learning_rate=0.005,alpha=0.001)
```

```python
Ridge.fit(X_train,y_train)
```

```python
y_pred = Ridge.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
