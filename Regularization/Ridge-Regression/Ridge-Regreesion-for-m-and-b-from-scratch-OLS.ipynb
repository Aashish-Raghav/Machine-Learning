```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=13)
```

```python
plt.scatter(X,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
print(lr.coef_,lr.intercept_)
```

```python
class MeraRidge:
    def __init__(self,alpha):
        self.alpha = alpha
        self.coef_ = None
        self.intercept_ = None

    def fit(self,X,y):
        num = 0
        den = 0
        for i in range(X.shape[0]):
            num = num + (y[i]- y.mean())*(X[i]-X.mean())
            den = den + (X[i]-X.mean())**2

        den = den + self.alpha
        self.coef_ = [num/den]
        self.intercept_ = y.mean()-self.coef_[0]*X.mean()
        print(self.coef_,self.intercept_)

    def predict(self,X):
        return np.dot(X,self.coef_) + self.intercept_
        
```

```python
Ridge = MeraRidge(0)
```

```python
Ridge.fit(X,y)
```

```python
Ridge.predict(X)
```

```python
alphas = [0,1,10,100,1000,10000]

plt.scatter(X,y)
Xnew = np.linspace(-3,3,100).reshape(100,1)
for alpha in alphas:
    Ridge = MeraRidge(alpha)
    Ridge.fit(X,y)
    plt.plot(Xnew,Ridge.predict(Xnew),label="Alpha {}".format(alpha))

plt.legend()
plt.show()
```
